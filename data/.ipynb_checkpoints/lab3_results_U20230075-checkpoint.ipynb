```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
```

load data

```python
train_users = pd.read_csv("train_users.csv")
test_users = pd.read_csv("test_users.csv")

train_users.head()
```

```python
train_users.info()
print(train_users.columns)
```

preprocessing

```python
X_train = train_users.drop("label", axis=1)
y_train = train_users["label"]
```

```python
X_test = test_users.copy()
```

```python
X_train = X_train.drop("user_id", axis=1)
X_test = X_test.drop("user_id", axis=1)
```

```python
X_train.fillna(X_train.mode().iloc[0], inplace=True)
X_test.fillna(X_test.mode().iloc[0], inplace=True)
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_train, X_test = X_train.align(
    X_test,
    join="left",
    axis=1,
    fill_value=0
)
```

traning

```python
model = DecisionTreeClassifier(
    max_depth=8,
    min_samples_split=10,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
test_predictions = model.predict(X_test)

test_predictions[:10]
```

```python
test_users_with_predictions = test_users.copy()
test_users_with_predictions["predicted_label"] = test_predictions

test_users_with_predictions.head()
```

```python
from sklearn.metrics import accuracy_score

train_predictions = model.predict(X_train)
train_accuracy = accuracy_score(y_train, train_predictions)

print("Training Accuracy:", train_accuracy)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    model,
    X_train,
    y_train,
    cv=5,
    scoring="accuracy"
)

print("Cross-validation accuracy:", cv_scores.mean())
```
